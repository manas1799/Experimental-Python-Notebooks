```python
import bs4
import pandas as pd
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
import pickle
import pandas as pd

import numpy as np
data = pd.read_csv('Desktop/BIOGRID-PUBLICATION-77183-4.0.189.mitab.txt',sep='tab')
data = np.array(data)
dictionary = {}
for i in range(data.shape[0]):
    
    one = ''
    two = ''
    if len(data[i][0].split('swiss-prot:'))>1:
        one = data[i][0].split('swiss-prot:')[1].split('|')[0]
    
    if len(data[i][0].split('swiss-prot:'))>2:
        two = data[i][0].split('swiss-prot:')[2].split('|')[0]
    
    if one not in dictionary.keys() and len(one)>1:
        my_url = f"https://www.uniprot.org/uniprot/{one}.fasta"
        try:
            uClient = uReq(my_url)
        except :
            print("File one not found")
            print("Seq was "+ str(one))
            continue
        page_html = str(uClient.read())
        #print(page_html.split("'")[1])
        #print(len(page_html.split("'")))
        if (len(page_html.split('GN='))<2 or len(page_html.split('GN=')[1].split()) <1 
            or len(page_html.split('\\n'))<1 ):
            print("File one not found")
            print("Seq was "+ str(one))
            continue
        #print(page_html.split('GN=')[1].split()[0])
        #print(len(page_html.split('GN=')))
        dictionary[page_html.split('GN=')[1].split()[0]] = ''.join(page_html.split('\\n')[1:])
    
    if two not in dictionary.keys() and len(two)>1:
        
        my_url = f"https://www.uniprot.org/uniprot/{two}.fasta"
        try:
            uClient = uReq(my_url)
        except :
            print("File two not found")
            print("Seq was "+ str(two))
            continue
        page_html = str(uClient.read())
        #print(page_html.split("'")[1])
        #print(page_html.split("'"))
        if (len(page_html.split('GN='))<2 or len(page_html.split('GN=')[1].split()) <1 
            or len(page_html.split('\\n'))<1 ):
            print("File one not found")
            print("Seq was "+ str(two))
            continue
        dictionary[page_html.split('GN=')[1].split()[0]] = ''.join(page_html.split('\\n')[1:])
    print("Iteration : " + str(i+1) + "/" + str(data.shape[0]))
    print("Dictionary size : " + str(len(dictionary)))


'''
with open('proteins_dictionary.pkl', 'wb') as fp:
    pickle.dump(dictionary, fp, protocol=pickle.HIGHEST_PROTOCOL)
    

with open('proteins_dictionary.pkl', 'rb') as fp:
    data = pickle.load(fp)    
    '''
    
print(data)
    
```
