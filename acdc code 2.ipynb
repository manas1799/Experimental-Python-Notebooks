```python
for i in range(1,100):
    if(i==90):
        continue
    if(i<10):
        ptx = nb.load("Desktop/ML/Datasets/training/patient00"+str(i)+"/patient00"+str(i)+"_frame01.nii")
        pty = nb.load("Desktop/ML/Datasets/training/patient00"+str(i)+"/patient00"+str(i)+"_frame01_gt.nii")
        print(i)
    elif(i>10 or i==10):
        ptx = nb.load("Desktop/ML/Datasets/training/patient0"+str(i)+"/patient0"+str(i)+"_frame01.nii")
        pty = nb.load("Desktop/ML/Datasets/training/patient0"+str(i)+"/patient0"+str(i)+"_frame01_gt.nii")
        print(i)
    
    
    ptn= np.asarray(ptx.dataobj)
    xi = ptn[:,:,0]
    for j in range(1,ptn.shape[2]):
        xi = np.vstack((xi,ptn[:,:,j]))
    
    
    xi=xi.reshape((ptn.shape[2],ptn.shape[0],ptn.shape[1]))
    np.save("NP_Pt_"+str(i),xi)
    
    
    ptn1= np.asarray(pty.dataobj)
    xi1 = ptn[:,:,0]
    for j in range(1,ptn1.shape[2]):
        xi1 = np.vstack((xi1,ptn[:,:,j]))
        
    
    xi1=xi1.reshape((ptn1.shape[2],ptn1.shape[0],ptn1.shape[1]))
    np.save("NP_Pt_gt_"+str(i),xi1)
    
```
